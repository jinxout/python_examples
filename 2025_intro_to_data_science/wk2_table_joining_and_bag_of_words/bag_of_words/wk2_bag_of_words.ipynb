```python
# Import the libraries
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

# (1) Create a Bag of Words model:
using "wordlist.txt" and **NOT** using stop words ("stop_words.txt")

```python
# open text file "wordlist.txt"
f = open("wordlist.txt", "r")
file_lines = f.readlines()
f.close()

file_lines
```

```python
# remove new line characters
word_list = [x[:-1] for x in file_lines]
word_list
```

```python
# Initiate the CountVectorizer class from scikit-learn
vectorizer = CountVectorizer()

# Learn the vocabulary from "word_list" and return document-term matrix
vectorizer.fit_transform(word_list)
```

```python
# Get the words in the vocabulary
words = vectorizer.get_feature_names_out()
words
```

# (2) Use the created Bag of Words model to generate a feature vector for:
”The patient is suffering from a migraine and is showing light sensitivity.” Find the words in the vocabulary.

```python
# Transform example text
note = "The patient is suffering from a migraine and is showing light sensitivity."
note_vector = vectorizer.transform([note]).toarray() # return document-term matrix

#Find indices of the words in the vocabulary
found_words = np.where(note_vector >= 1)[1].tolist()
found_words
```

```python
# Find the words in the vocabulary
for i in found_words:
  print(words[i])
```

# (3) Create a Bag of Words model:
using wordlist.txt **and using stop words ("stop_words.txt")**.

```python
! ls
```

```python
# open text file "stop_words.txt" and remove new line characters
f = open("stop_words.txt", "r")
file_lines_stop = f
stop_list = [x[:-1] for x in file_lines_stop]
f.close()

stop_list
```

```python
# Initiate the CountVectorizer class from scikit-learn, assign "stop_list" to stop_words parameter.
vectorizer_stopped = CountVectorizer(stop_words=stop_list)

# Create vocabulary using "word_list" and the "stop_list".
vectorizer_stopped.fit_transform(word_list)
```

```python
# Get the words in the vocabulary that does not have stop words
words_stopped = vectorizer_stopped.get_feature_names_out()
words_stopped
```
