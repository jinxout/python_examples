```python
# Import the libraries
import pandas as pd
import numpy as np
```

# (1) Loads CSVs:
using the data from "breast-cancer-wisconsin-left.csv" as the left table and the data from "breast-cancer-wisconsin-right.csv" as the right table.

```python
df_left = pd.read_csv("breast-cancer-wisconsin-left.csv", header=0)
df_left
```

```python
df_right = pd.read_csv("breast-cancer-wisconsin-right.csv", header=0)
df_right
```

# (2) Performs an inner join.

```python
df_inner = df_left.merge(df_right, on='ID Number', how='inner')
df_inner
```

# (3) Performs a left join.

```python
df_left = df_left.merge(df_right, on='ID Number', how='left')
df_left
```

# (4) Performs a right join.

```python
df_right = df_left.merge(df_right, on='ID Number', how='right')
df_right
```

# (5) Performs an outer join.

```python
df_outer = df_left.merge(df_right, on='ID Number', how='outer')
df_outer
```
